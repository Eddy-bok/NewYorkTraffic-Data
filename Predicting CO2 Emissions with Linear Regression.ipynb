#### Step 1: Import Necessary Packages

```python
# import necessary python packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
%matplotlib inline
```

#### Step 2: Load and Explore the Dataset

```python
# Load and Explore the dataset
def load_data(file_name):
    data = pd.read_csv(file_name)
    return data["engine_size"].values, data["co2_emissions"].values
```

###### Step 2: Analysis

```python
x_train, y_train = load_data("D:/Data Analytics Courses/3rd Semester - Spring 2025/Course Materials/DAT 514/datasets/co2_emissions.csv")
print("First 5 entries of x_train:", x_train[:5])
print("First 5 entries of y_train:", y_train[:5])
```

#### Step 3: Visualize the Data

```python
# Visualize the Data
# Create a scatter plot to visualize the relationship between engine size and C02 emissions
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("C02 Emissions vs Engine Size")
plt.xlabel("Engine Size (liters)")
plt.ylabel("C02 Emissions (g/km)")
plt.show()
```

###### Step 3: Analysis

#### Step 4: Compute the Cost Function

```python
def compute_cost(x, y, w, b):
    m = len(x)
    total_cost = 0
    for i in range(m):
        total_cost += (w * x[i] + b - y[i]) ** 2
    return total_cost / (2 * m)
```

###### Step 4: Analysis

#### Step 5: Gradient Descent

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters):
    m = len(x)
    w, b = w_in, b_in
    for _ in range(num_iters):
        dj_dw, dj_db = 0, 0
        for i in range(m):
            dj_dw += (w * x[i] + b - y[i]) * x[i]
            dj_db += (w * x[i] + b - y[i])
        w -= alpha * dj_dw / m
        b -= alpha * dj_db / m
    return w, b
```

```python
w, b = gradient_descent(x_train, y_train, w_in=0, b_in=0, alpha=0.01, num_iters=1500)
print("Optimized w: ", w)
print("Optimized b: ", b)
```

###### Step 5: Analysis

#### Step 6: Predict CO2 Emissions

```python

def predict(x, w, b):
    return w * x + b
# Example Predictions
engine_sizes = [1.5, 3.0]
for size in engine_sizes:
    print(f"For engine size {size}L, predicted CO2 emissions: {predict(size, w, b):.2f} g/km")
```

###### Step 6: Analysis

#### Diagonistic Graph Showing Actual VS Predicted Values

```python
# Predictions for the training data
predictions = [predict(x, w, b) for x in x_train]

plt.scatter(x_train, y_train, color='red', marker='x', label='Actual')
plt.plot(x_train, predictions, color='blue', label='Predicted')
plt.title("CO2 Emissions vs Engine Size")
plt.xlabel("Engine Size (liters)")
plt.ylabel("CO2 Emissions (g/km)")
plt.legend()
plt.show()
```

#### Residual Plot

```python
residuals = [y - predict(x, w, b) for x, y in zip(x_train, y_train)]

plt.scatter(x_train, residuals, color='green')
plt.axhline(0, color='black', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Engine Size (liters)")
plt.ylabel("Residuals")
plt.show()
```
